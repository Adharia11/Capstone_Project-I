```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Electric_Vehicle_Population_Data.csv')
```

```python
print(data.head())
print(data.info())
```

```python
ev_counts_by_year = data.groupby('Model Year')['Make'].count().reset_index()
print(ev_counts_by_year)
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='Model Year', y='Make', data=ev_counts_by_year)
plt.xlabel('Model Year')
plt.ylabel('Electric Vehicle Count')
plt.title('Electric Vehicle Population Over the Years')
plt.grid(True)
plt.show()
```

```python
ev_count_by_state = data.groupby('State')['Make'].count().reset_index()
print (ev_count_by_state)

```

```python
# Plot the distribution of electric vehicles across regions
plt.figure(figsize=(12, 6))
sns.barplot(x='State', y='Make', data=ev_count_by_state)
plt.xlabel('State')
plt.ylabel('Electric Vehicle Count')
plt.title('Electric Vehicle Population by Region')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
if 'Base MSRP' in data.columns:
    most_expensive_ev = data.sort_values(by='Base MSRP', ascending=False).head(20)

print (most_expensive_ev)


```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='Base MSRP', y='Make', data=most_expensive_ev)
plt.title('Electric Vehicle Adoption Based on Cost')
plt.xlabel('Base MSRP')
plt.ylabel('Number of Electric Vehicles')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
